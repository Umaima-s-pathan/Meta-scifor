```python
import numpy as np
df1=pd.DataFrame({'a':[1,2,3],'s':[4,np.nan,7],'h':['joshi',np.nan,np.nan],'f':[np.nan,7,'firoz']})
df1  # inserting null values in dataframe
```

```python
df1['g']=[20,30,40]
df1
```

```python
print("\n Dropping any column with null values\n",'-'*35,sep='')
print(df1.dropna(axis=1))  # here we are dropping  the null values column wise
```

```python
df1  # orignal dataframe is as it is
```

```python
print("\n Dropping any column with null values\n",'-'*35,sep='')
print(df1.dropna(axis=0))   # here we are dropping  the null values row wise
# since we have null values are every row that is why it has droped all the rows and now the dataframe is empty
```

```python
df1 # but the original dataframe is as it  is
```

```python
print("\n Filling Value with a Default Value\n",'-'*35,sep='')
print(df1.fillna(value=10))  # filling the null value as 10
```

```python
df1
```

```python
print("\n Filling Values with a computed values (mean of a column A here)\n",'-'*70,sep='')
# using inplacewe can change the values of the datframe perminantly default vale of inplace is false
df1.fillna(value=df1['a'].mean(),inplace=True) # mentioning inplace = true to make the changes perminant
print(df1)
```

```python
df1  #now the null values are perminantly changed
```

```python
data={'company':['JOP','FTY','FTY','FTY','GFB','JOP'],'person':['raha','arun','raj','sid','rauf','nabi'],'sales':[806,549,847,353,251,954]}
df2=pd.DataFrame(data)
df2
```

```python
class1=pd.DataFrame({'Mat':[45,68,35,65],
                     'Eng':[54,87,36,52],
                     'Sci':[66,77,88,70],
                     'Chem':[54,79,58,76]},
                    index=['ravi','shyam','saurav','abhi'])
class1
```

```python
class2=pd.DataFrame({'Mat':[65,32,44,48],
                     'Eng':[54,87,78,52],
                     'Sci':[65,68,88,95],
                     'Chem':[67,85,38,95]},
                    ['rauf','aadi','hasan','liz'])
class2
```

```python
class3=pd.DataFrame({'Mat':[86,46,90,65],
                 'Eng':[96,35,78,59],
                     'Sci':[94,38,84,93],
                     'Chem':[94,67,85,67]},
                    index=['ayesh','hatim','zoya','surbhi'])
class3
```

```python
# using concatination method to merge all the above 3 dataframe
asm=pd.concat([class1,class2,class3],axis=0)  # concatinating row wise
print("\n After concatenation along row \n",'-'*30,sep='')
print(asm)
```

```python
asm1=pd.concat([class1,class2,class3],axis=1)  # concatinating column wise
print("\n After concatenation along column \n",'-'*30,sep='')
print(asm1)
```

```python
asm1.fillna(value=0,inplace=True)
asm1
```

```python
df3=pd.DataFrame({'names':['ayesh','hatim','zoya','surbhi'],
                 'Age':[36,34,20,28]})
df3
```
